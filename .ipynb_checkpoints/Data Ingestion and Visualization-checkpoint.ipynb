```python
# from https://matplotlib.org/examples/pylab_examples/histogram_percent_demo.html
def to_percent(y, position):
    # Ignore the passed in position. This has the effect of scaling the default
    # tick locations.
    s = str(int(100 * y)) # added int to round numbers

    # The percent symbol needs escaping in latex
    if matplotlib.rcParams['text.usetex'] is True:
        return s + r'$\%$'
    else:
        return s + '%'
```

```python
X = df['SLOGAN']
y = df['SectorId']
```

```python
# vocabulary mostly used in slogans
import yellowbrick as yb
from yellowbrick.text.freqdist import FreqDistVisualizer
# Visualize frequency distribution of top xx tokens
vectorizer = CountVectorizer(ngram_range=(1, 2), max_df = 0.6,stop_words='english')
docs = vectorizer.fit_transform(X)
features = vectorizer.get_feature_names()

visualizer = FreqDistVisualizer(features,n=20)
visualizer.fit(docs)
visualizer.poof()
# total number of words: 2,877
# number of different words: 1,941
# number of words used once:1,610
```

```python
import string
# remove the punctuation
pos = []
for x in X:
    for punc in string.punctuation:
        x = x.replace(punc,'')
    pos.append(x)
# From ['Science. Applied to life.' .... 'Life. To the fullest.'  ......]    
# To   ['Science Applied to life',.....'Life To the fullest',.....]
```

```python
from nltk.tokenize import word_tokenize
vocabulary = []
for x in pos:
    vocabulary.append(word_tokenize(x)) # Create on big list of sentence 
#  [['Science', 'Applied', 'to', 'life'],..., ['Life', 'To', 'the', 'fullest'],.....] 
```

```python
bagofwords = sum(vocabulary,[]) # one list with all the elements tokenized
# ['Science','Applied','to','life','At','the','AP','we','watch','our','Ps','and','Qs','Life','To','the','fullest',...]
```

```python
slogans_tagged = nltk.pos_tag(bagofwords)
# [('Science', 'NN'),('Applied', 'NNP'),('to', 'TO'),('life', 'NN'),('At', 'IN'),('the', 'DT'),('AP', 'NNP'), .....]
```

```python
tag_fd = nltk.FreqDist(tag for (word, tag) in slogans_tagged)
```

```python
import itertools
from nltk.tokenize import sent_tokenize
key_words = ['world','life','energy','people','good','better','way','make','lives','company','think',
             'business','power','home','things','future','work','I','new','everyday']
```

```python
possible_pairs = list(itertools.combinations(key_words, 2))
cooccurring = dict.fromkeys(possible_pairs, 0)

for sent in vocabulary:
    for pair in possible_pairs:
        if pair[0] in sent and pair[1] in sent:
            cooccurring[pair] += 1

# cooccurring
#{('better', 'everyday'): 1,....('better', 'work'): 1,('business', 'future'): 0, ('make', 'things'): 2}
```

```python
import networkx as nx
G = nx.Graph()
G.name = "The social network of Slogans"
for pair,fois in cooccurring.items():
    if fois > 0:
        G.add_edge(pair[0], pair[1], weight=fois)
        
D = nx.ego_graph(G, "life")
edges, weights = zip(*nx.get_edge_attributes(D, "weight").items())
# Push nodes away that are less related to life
pos = nx.spring_layout(D, k=.5, iterations=40)
nx.draw(D, pos, node_color="gold", node_size=50, edgelist=edges,
         width=.5, edge_color="orange", with_labels=True, font_size=12)
plt.title("The Social Network of Slogans",fontsize=20)
plt.show()
```
