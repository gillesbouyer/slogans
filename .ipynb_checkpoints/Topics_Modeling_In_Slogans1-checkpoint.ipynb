```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(input ='content',
                                   stop_words='english', 
                                   token_pattern="\\b[a-z][a-z]+\\b",
                                   lowercase=True)
from sklearn.decomposition import LatentDirichletAllocation
import numpy as np

lda = LatentDirichletAllocation(n_components = 3)
```

```python
!pwd
```

```python
gen = (record['SLOGAN'] for record in slogans_collection.find())
print(gen)
cv.fit(gen)
gen = (record['SLOGAN'] for record in slogans_collection.find())
#gen = (record['SLOGAN'] for record in slogans_collection.find())
for file in gen:
    vec_file = cv.transform([file])
    lda.partial_fit(vec_file)
```

```python
def display_topics(model, feature_names, no_top_words, topic_names=None):
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix)
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
        
display_topics(lda, cv.get_feature_names(), 10)
```

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import sklearn
# Import all of the scikit learn stuff
from __future__ import print_function
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.preprocessing import Normalizer
from sklearn import metrics
from sklearn.cluster import KMeans, MiniBatchKMeans
import pandas as pd

#%matplotlib inline

import sys
import nltk
```

```python
df = pd.read_excel('./data/aug18slogan.xlsx')
```

```python
# As entry of LSA we want: example = ["Machine learning is super fun", 
#            "Python is super, super cool",
#             "Statistics is cool, too",
X = df['SLOGAN']
y = df['SectorId']
import string
# remove the punctuation
pos = []
for x in X:
    for punc in string.punctuation:
        x = x.replace(punc,'')
    pos.append(x)
# From ['Science. Applied to life.' .... 'Life. To the fullest.'  ......]    
# To   ['Science Applied to life',.....'Life To the fullest',.....]   ---> ok for LSA
```

```python
vectorizer = CountVectorizer(min_df = 1, stop_words = 'english')
dtm = vectorizer.fit_transform(pos) 
pd.DataFrame(dtm.toarray(), index=pos, columns=vectorizer.get_feature_names()).head(10)
```

```python
# Get words that correspond to each column 
vectorizer.get_feature_names()
```

```python
# Fit LSA. Use algorithm = “randomized” for large datasets 
from sklearn.decomposition import TruncatedSVD
lsa = TruncatedSVD(2, algorithm = 'arpack')
dtm_lsa = lsa.fit_transform(dtm.asfptype())
dtm_lsa = Normalizer(copy=False).fit_transform(dtm_lsa)
```

```python
lsa.explained_variance_ratio_
```

```python
pd.DataFrame(lsa.components_.round(5),index = ["component_1","component_2"],columns = vectorizer.get_feature_names())
```

```python
# let's check what words make component_1 / component_2 react the most
```

```python
df3 = pd.DataFrame(dtm_lsa.round(5), index = pos, columns = ["component_1","component_2" ])
```

```python
df3.nlargest(20, 'component_1')
```

```python
df3.nlargest(20, 'component_2')
```

```python
# NMF
from sklearn.decomposition import NMF

nmf_model = NMF(2)

dtm_nmf = nmf_model.fit_transform(dtm)
dtm_nmf = Normalizer(copy=False).fit_transform(dtm_nmf)
```
